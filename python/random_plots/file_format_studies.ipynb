```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib notebook
```

```python
filetypes = ['ROOT','Text','Zipped text','HDF5','HDF5 compressed']

files = []
files.append({'name':'data.X','ROOTsize':14403501, 'ROOTspeed':32423423,
                           'Textsize':32619674, 'Textspeed':32423423,
                           'Zipsize':12168467, 'Zipspeed':32423423,
                           'HDF5size':30677896,'HDFspeed':324324,
                           'HDF5compressedsize':8362708,'HDFcompressedspeed':324324})

#             ttbar.X','ww.X']

```

```python
files
```

```python
keys = files[0].keys()
sizekeys = []
for k in keys:
    if 'size' in k:
        sizekeys.append(k)
```

```python
sizes = []
for k in sizekeys:
    sizes.append(files[0][k]/1e6)
print(sizes)
```

```python
plt.figure()
x = np.arange(0,len(sizes))
plt.xticks(x, filetypes)
plt.plot(x,sizes,'o',markersize=10)
plt.ylim(0,1.2*max(sizes))
plt.ylabel('Filesize (Mb)',fontsize=18)
plt.tight_layout()
```
