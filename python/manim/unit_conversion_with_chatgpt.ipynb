```python
%%manim -qm UnitConversion
class UnitConversion(Scene):
    def construct(self):
        # Create the initial equation "1 = 1"
        eq = MathTex("1", "=", "1")
        # Fade in the entire equation over 2 seconds.
        self.play(FadeIn(eq), run_time=2)
        self.wait(0.5)

        # --- First transformation: Change the right-hand side "1" to "2/2" ---
        frac_2_2 = MathTex(r"\frac{2}{2}")
        # Place the new fraction exactly where the old right-hand side was.
        frac_2_2.move_to(eq[2])
        # Transform the right-hand side into the new fraction.
        self.play(ReplacementTransform(eq[2], frac_2_2), run_time=1)
        # Update our equation's right-hand side reference.
        eq[2] = frac_2_2
        self.wait(0.5)

        # --- Second transformation: Change "2/2" to "5/5" ---
        frac_5_5 = MathTex(r"\frac{5}{5}")
        frac_5_5.move_to(eq[2])
        self.play(ReplacementTransform(eq[2], frac_5_5), run_time=1)
        eq[2] = frac_5_5
        self.wait(0.5)

        # --- Third transformation: Change "5/5" to "1000000/1000000" ---
        frac_1000000 = MathTex(r"\frac{1000000}{1000000}")
        frac_1000000.move_to(eq[2])
        self.play(ReplacementTransform(eq[2], frac_1000000), run_time=1)
        eq[2] = frac_1000000
        self.wait(1)
```

```python
%%manim -qm UnitConversion

class UnitConversion(Scene):
    def construct(self):
        # Create the initial equation "1 = 1"
        eq = MathTex("1", "=", "1")
        self.play(FadeIn(eq), run_time=2)
        self.wait(0.5)

        # Use a separate variable for the right-hand side (rhs)
        rhs = eq[2]

        # --- First transformation: "1" -> "2/2" ---
        frac_2_2 = MathTex(r"\frac{2}{2}")
        frac_2_2.move_to(rhs)
        self.play(ReplacementTransform(rhs, frac_2_2), run_time=1)
        rhs = frac_2_2
        self.wait(0.5)

        # --- Second transformation: "2/2" -> "5/5" ---
        frac_5_5 = MathTex(r"\frac{5}{5}")
        frac_5_5.move_to(rhs)
        self.play(ReplacementTransform(rhs, frac_5_5), run_time=1)
        rhs = frac_5_5
        self.wait(0.5)

        # --- Third transformation: "5/5" -> "1000000/1000000" ---
        # Save the original left edge position of the rhs so we can align to it.
        original_rhs_left = rhs.get_left()

        frac_1000000 = MathTex(r"\frac{1000000}{1000000}")
        # Shift frac_1000000 so that its left edge matches the original rhs left edge.
        frac_1000000.shift(original_rhs_left - frac_1000000.get_left())
        # Animate the transformation.
        self.play(ReplacementTransform(rhs, frac_1000000), run_time=1)
        rhs = frac_1000000
        self.wait(1)
```

```python
%%manim -qm IntegrationByPartsDemo

class IntegrationByPartsDemo(Scene):
    def construct(self):
        # EXAMPLE EQUATION #1
        eq1 = MathTex(
            r"\int u(x)\,v'(x)\,dx = u(x)v(x) - \int u'(x)v(x)\,dx"
        )
        # EXAMPLE TEXT #1
        text1 = Tex(
            "We start by considering two continuously differentiable functions "
            r"$u(x)$ and $v(x)$. By applying the product rule and integrating, "
            "we arrive at the standard formula for integration by parts."
        ).scale(0.6)  # Just scaling text a bit so it fits nicely

        # Step 1: Show the first equation
        self.play(Write(eq1))
        # Step 2: Wait for 3 seconds
        self.wait(3)

        # Step 3: Move the equation up to make space for text
        self.play(eq1.animate.to_edge(UP))
        self.wait(0.5)

        # Step 4: Reveal the text below the equation
        text1.next_to(eq1, DOWN, buff=0.5)  # place text below eq1
        self.play(FadeIn(text1))
        self.wait(3)

        # Step 5: Clear the screen (fade out equation + text)
        self.play(FadeOut(eq1), FadeOut(text1))
        self.wait(1)

        # EXAMPLE EQUATION #2
        eq2 = MathTex(
            r"\int u\,dv = uv - \int v\,du"
        )
        text2 = Tex(
            "This is often written in a shorter differential form, and "
            "it's useful when $u$ and $dv$ are chosen for convenience."
        ).scale(0.6)

        # Show second equation
        self.play(Write(eq2))
        self.wait(3)
        # Move it up
        self.play(eq2.animate.to_edge(UP))
        self.wait(0.5)
        # Reveal text
        text2.next_to(eq2, DOWN, buff=0.5)
        self.play(FadeIn(text2))
        self.wait(3)
        # Clear
        self.play(FadeOut(eq2), FadeOut(text2))
        self.wait(1)

        # ... repeat for additional equations/text ...
```

```python
%%manim -qm IntegrationByPartsVideo

class IntegrationByPartsVideo(Scene):
    def construct(self):
        # --- 1) (u(x)v(x))' = u'(x)v(x) + u(x)v'(x) ---
        eq1 = MathTex(r"(u(x)v(x))' = u'(x)v(x) + u(x)v'(x)")
        text1 = Tex(
            r"The theorem can be derived as follows. "
            r"For two continuously differentiable functions $u(x)$ and $v(x)$, "
            r"the product rule states:"
        ).scale(0.6)

        self.play(Write(eq1))
        self.wait(3)
        self.play(eq1.animate.to_edge(UP))
        self.wait(0.5)
        text1.next_to(eq1, DOWN, buff=0.5)
        self.play(FadeIn(text1))
        self.wait(3)
        self.play(FadeOut(eq1), FadeOut(text1))
        self.wait(1)

        # --- 2) ∫(u'(x)v(x) + u(x)v'(x)) dx = ∫(u(x)v(x))' dx ---
        eq2 = MathTex(
            r"\int \bigl(u'(x)v(x) + u(x)v'(x)\bigr)\,dx \;=\;"
            r"\int \bigl(u(x)v(x)\bigr)'\,dx"
        )
        text2 = Tex(
            r"Integrating both sides with respect to $x$, we obtain:"
        ).scale(0.6)

        self.play(Write(eq2))
        self.wait(3)
        self.play(eq2.animate.to_edge(UP))
        self.wait(0.5)
        text2.next_to(eq2, DOWN, buff=0.5)
        self.play(FadeIn(text2))
        self.wait(3)
        self.play(FadeOut(eq2), FadeOut(text2))
        self.wait(1)

        # --- 3) ∫ u'(x)v(x) dx + ∫ u(x)v'(x) dx = u(x)v(x) + C ---
        eq3 = MathTex(
            r"\int u'(x)v(x)\,dx \;+\; \int u(x)v'(x)\,dx \;=\; u(x)v(x) + C"
        )
        text3 = Tex(
            r"Given that an indefinite integral is an antiderivative, "
            r"we can write:"
        ).scale(0.6)

        self.play(Write(eq3))
        self.wait(3)
        self.play(eq3.animate.to_edge(UP))
        self.wait(0.5)
        text3.next_to(eq3, DOWN, buff=0.5)
        self.play(FadeIn(text3))
        self.wait(3)
        self.play(FadeOut(eq3), FadeOut(text3))
        self.wait(1)

        # --- 4) ∫ u(x)v'(x) dx = u(x)v(x) - ∫ u'(x)v(x) dx ---
        eq4 = MathTex(
            r"\int u(x)\,v'(x)\,dx \;=\; u(x)v(x) \;-\; \int u'(x)\,v(x)\,dx"
        )
        text4 = Tex(
            r"Where we neglect writing the constant of integration. "
            r"This yields the formula for integration by parts:"
        ).scale(0.6)

        self.play(Write(eq4))
        self.wait(3)
        self.play(eq4.animate.to_edge(UP))
        self.wait(0.5)
        text4.next_to(eq4, DOWN, buff=0.5)
        self.play(FadeIn(text4))
        self.wait(3)
        self.play(FadeOut(eq4), FadeOut(text4))
        self.wait(1)

        # --- 5) ∫ u dv = uv - ∫ v du ---
        eq5 = MathTex(r"\int u\,dv \;=\; uv \;-\; \int v\,du")
        text5 = Tex(
            r"Or in terms of the differentials $du$ and $dv$:"
        ).scale(0.6)

        self.play(Write(eq5))
        self.wait(3)
        self.play(eq5.animate.to_edge(UP))
        self.wait(0.5)
        text5.next_to(eq5, DOWN, buff=0.5)
        self.play(FadeIn(text5))
        self.wait(3)
        self.play(FadeOut(eq5), FadeOut(text5))
        self.wait(1)

        # --- 6) ∫[a->b] u(x)v'(x) dx = [u(x)v(x)]_a^b - ∫[a->b] u'(x)v(x) dx ---
        eq6 = MathTex(
            r"\int_a^b u(x)\,v'(x)\,dx \;=\; \bigl[u(x)v(x)\bigr]_a^b"
            r"\;-\; \int_a^b u'(x)\,v(x)\,dx"
        )
        text6 = Tex(
            r"This is to be understood as an equality of functions with an "
            r"unspecified constant added to each side. Taking the difference "
            r"of each side between two values $x = a$ and $x = b$ and applying "
            r"the fundamental theorem of calculus gives the definite integral "
            r"version:"
        ).scale(0.6)

        self.play(Write(eq6))
        self.wait(3)
        self.play(eq6.animate.to_edge(UP))
        self.wait(0.5)
        text6.next_to(eq6, DOWN, buff=0.5)
        self.play(FadeIn(text6))
        self.wait(3)
        self.play(FadeOut(eq6), FadeOut(text6))
        self.wait(1)

        # --- 7) Final text only (no new equation) ---
        final_text = Tex(
            r"The original integral $\int u\,dv$ contains the derivative $v'$. "
            r"To apply the theorem, one must find $v$, the antiderivative of $v'$, "
            r"then evaluate the resulting integral $\int v\,du$."
        ).scale(0.6)

        self.play(FadeIn(final_text))
        self.wait(3)
        self.play(FadeOut(final_text))
        self.wait(1)
```
