# Loop over just the strains

```python
# Get all the values

exptypes = df['ExpType'].values
fatypes = df['FAType'].values
strains = df['Strain'].values
mM = df['mM'].values
mM2 = df['mM2'].values
pg = df['pg'].values

# Concentration holder
conc = {}

for m in mM:
#for m in mM2:

    conc[m] = []

includes = df['Include']


exptype = 'Biofilm'
fatype = 'dec'

for exptype in np.unique(exptypes):
    for fatype in np.unique(fatypes):

        idx = (exptypes==exptype)*(fatypes==fatype)*(includes=='Yes')

        colors = ['b','g','r','c','m','y','k','b','g','r','c']

        strain_plots = {}
        #'''


        plot_data = {}
        for i,strain_label in enumerate(strain_labels[:]):

            plt.figure(figsize=(4,4))
            plt.title(strain_label)
            #print(strain_label)

            for key in conc.keys():
                conc[key] = []

                idx2 = idx*(strains==strain_label)

                y = pg[idx2]
                mMvals = mM[idx2]
                #mMvals = mM2[idx2].astype(str) # These are strings
                #print(mMvals,y)
                for a,b in zip(mMvals,y):
                    conc[a].append(b)

                plt.plot(mMvals,y,'.',color=colors[i])

            x = []
            y = []
            yerr = []
            #print(conc)
            for key in conc.keys():
                y.append(np.mean(conc[key]))
                yerr.append(np.std(conc[key]))
                x.append(key)
            #print(x,y,yerr)
            plt.errorbar(x,y,yerr,fmt='o')
            plot_data[strain_label] = [x,y,yerr]

            plt.xscale('log')


        ## Now plot them all
        #xlabels = []
        #for m in mM:
        #    concentrations[i] = concentrations[i-1]/2.
        #    xlabels.append('{0:2.3f} mM'.format(concentrations[i]))





        plt.figure(figsize=(8,8))

        icount = 0
        for key in np.sort(list(plot_data.keys())):
            x,y,yerr = plot_data[key]
            print(x)
            plt.subplot(3,3,icount+1)
            plt.errorbar(x,y,yerr,fmt='o')
            plt.title(key)
            plt.xscale('log')
            plt.ylim(0,200)
            plt.xlabel('mM')
            plt.ylabel('% growth')

            plt.gca().get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
            plt.gca().get_xaxis().set_minor_formatter(matplotlib.ticker.NullFormatter())

            plt.tight_layout()
            icount += 1
        filename = 'summary_plot_{0}_{1}.png'.format(exptype,fatype)
        plt.savefig(filename)


        #''';
```
