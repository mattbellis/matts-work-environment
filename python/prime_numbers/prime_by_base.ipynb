```python
import math
import numpy as np
import matplotlib.pylab as plt

from is_prime import is_prime, get_factors
```

```python
def convert_base(n, base=10, verbose=False):
    digits = []
    for i in range(10):
        divisor =      int(math.pow(base,i))
        divisor_plus = int(math.pow(base,i+1))

        digit = int((n  % divisor_plus) / divisor)

        digits.append(digit)
        
        if verbose:
            print(f'{n}   {digit}   {divisor:12d} {divisor_plus:12d}')

        if divisor_plus >= n:
            break
            
    return digits


digits = convert_base(453, 8, verbose=True)

print(digits)
```
