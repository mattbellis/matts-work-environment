```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import seaborn as sn
import lichen.lichen as lch
```

```python
infilename = 'SAT_scores_2014.csv'
term,math,verbal,total,school,major = np.loadtxt(infilename,skiprows=8,unpack=True,dtype=str,delimiter=',')

term[term==''] = '0'
math[math==''] = '0'
verbal[verbal==''] = '0'
total[total==''] = '0'

math=math.astype('float')
verbal=verbal.astype('float')
total=total.astype('float')
term=term.astype(int)/100

schools = ['SD','AD','BD']
colors = ['blue','red','green']

allyears = range(2007,2017)
```

```python
means = {'SD':[],'AD':[],'BD':[]}
stds = {'SD':[],'AD':[],'BD':[]}



plt.figure(figsize=(24,16))
for i in allyears:
    values = total
    index = term==i
    index *= values!=0
    indexes = [index*(school=='SD'),index*(school=='AD'),index*(school=='BD')]

    plt.subplot(4,4,i+1-2007)
    for s,idx,c in zip(schools,indexes,colors):
        name = "%s %d" % (s,i)
        plt.hist(values[idx],bins=25,range=(400,1600),label=name,histtype='stepfilled',alpha=0.5,color=c)
        mean = np.mean(values[idx])
        std = np.std(values[idx])
        
        means[s].append(mean)
        stds[s].append(std)
        
        #print "%s %f %f" % (name,mean,std)
        plt.plot([mean,mean],[0,40],color=c)
        plt.xlim(400,1700)
        plt.ylim(0,40)
    plt.legend()

import matplotlib.dates as mdates

syears = []
for y in allyears:
    syears.append("1/1/%d" % (y))
dyears = mdates.datestr2num(syears)
for d in dyears:
    print d

plt.figure(figsize=(24,10))
for s,c in zip(schools,colors):
    plt.subplot(1,2,1)
    plt.plot(dyears,means[s],'o-',color=c,label=s)
    plt.subplot(1,2,2)
    plt.plot(dyears,stds[s],'o-',color=c,label=s)

    
# format the ticks
plt.gcf().autofmt_xdate()

print allyears
years    = mdates.YearLocator()
yearsFmt = mdates.DateFormatter('%Y')

for i in [1,2]:
    plt.subplot(1,2,i)
    plt.gca().xaxis.set_major_locator(years)
    plt.gca().xaxis.set_major_formatter(yearsFmt)
    plt.gca().format_xdata = mdates.DateFormatter('%Y')
    plt.legend()
```

```python
nums = {'SD':[],'AD':[],'BD':[]}
numshi = {'SD':[],'AD':[],'BD':[]}
numslo = {'SD':[],'AD':[],'BD':[]}

hilo = [1300,1000]

plt.figure(figsize=(32,24))
for i in allyears:
    values = total
    index = term==i
    index *= values!=0
    indexes = [index*(school=='SD'),index*(school=='AD'),index*(school=='BD')]

    for s,idx,c in zip(schools,indexes,colors):
        name = "%s %d" % (s,i)

        nums[s].append(len(values[idx]))

        
        idxlo = idx*(values<hilo[1])
        numslo[s].append(len(values[idxlo]))

        idxhi = idx*(values>hilo[0])
        numshi[s].append(len(values[idxhi]))
 

plt.figure(figsize=(24,6))
for s,c in zip(schools,colors):
    plt.subplot(1,3,1)
    plt.plot(dyears,nums[s],'o-',color=c,label=s)
    plt.subplot(1,3,2)
    plt.plot(dyears,numshi[s],'o-',color=c,label=s)
    plt.subplot(1,3,3)
    plt.plot(dyears,numslo[s],'o-',color=c,label=s)

    
# format the ticks
plt.gcf().autofmt_xdate()

print allyears
years    = mdates.YearLocator()
yearsFmt = mdates.DateFormatter('%Y')

for i in [1,2,3]:
    plt.subplot(1,3,i)
    plt.gca().xaxis.set_major_locator(years)
    plt.gca().xaxis.set_major_formatter(yearsFmt)
    plt.gca().format_xdata = mdates.DateFormatter('%Y')
    plt.legend()

    if i==2:
        title = "# with SAT>%d" % (hilo[0])
    elif i==3:
        title = "# with SAT<%d" % (hilo[1])
    elif i==1:
        title = "Total # of students by school"

    plt.title(title,fontsize=24)
    
```
