$$y_1 = N_0 e^{-k x_1}$$
$$y_2 = N_0 e^{-k x_2}$$

$$(y_1/y_2) = e^{-k x_1}/e^{-x k_2}$$
$$(y_1/y_2) = e^{-k (x_2 - x_1)}$$

$$\ln (y_1/y_2) = -k (x_2 - x_1)$$
$$-\ln (y_1/y_2)/(x_2-x_1) = k$$
