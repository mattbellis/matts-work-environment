```python
infilename = 'LocationPhysicsDept.csv'

lats = []
lons = []
schools = []
enrollments = []
types = [[],[],[],[],[]]

infile = open(infilename,'r')

vals = np.loadtxt(infilename,dtype=str,unpack=True,skiprows=1,delimiter=',')

schools = vals[1]
for v in vals[6]:
    lats.append(float(v.replace('\"','')))

for v in vals[7]:
    lons.append(float(v.replace('\"','')))

for v in vals[8]:
    enrollments.append(float(v.replace('\"','')))

for i in range(0,5):
    for v in vals[9+i]:
        types[i].append(int(v.replace('\"','')))
        
types = np.array(types).transpose()
```

```python
max_enrollment = max(enrollments)
min_enrollment = min(enrollments)

print(max_enrollment, min_enrollment)

types
```

```python
def popup(school,enrollment,t):

    html="""
        <b>{0}</b><br><br>
        Enrollment: {1}<br><br>
    """.format(school,int(enrollment))
    
    if t[0]==1:
        html += "Certificate of less than one academic year in Physical Sciences<br>"
    
    if t[1]==1:
        html += "Certificate of at least one but less than two academic years in Physical Sciences.<br>"
        
    if t[2]==1:
        html += "Associate degree in Physical Sciences<br>"
        
    if t[3]==1:
        html += "Award of more than two but less than four academic years in Physical Sciences<br>"
        
    if t[4]==1:
        html += "Bachelor degree in Physical Sciences<br>"
    
    
    html += '<p>'
    iframe = folium.IFrame(html=html, width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    
    return popup
```
