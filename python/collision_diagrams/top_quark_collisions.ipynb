```python
###############################################################################
def jet(angle,mag):
    nlines = np.random.randint(2,6)
    angles = angle + 0.3*np.random.random(nlines)-0.25
    mags = mag + (0.5*np.random.random(nlines) - 0.25)

    return angles,mags
################################################################################
```

```python
################################################################################
def draw_beam():

    plt.gca()    
    # x, y, dx, dy
    plt.gca().arrow(-0.9, 0, 0.8, 0, head_width=0.05, head_length=0.1, fc='k', ec='k')
    plt.gca().arrow(0.9, 0, -0.8, 0, head_width=0.05, head_length=0.1, fc='k', ec='k')

################################################################################
```
